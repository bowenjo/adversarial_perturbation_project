```python
# learning rate
eta=1e-4

# number of trials 
numTrials = 200

# number of examples and batch size per trial
numExamples = 50000
batchSize = 100
numInputUnits = 784
numOutputUnits = 10
numSamples = int(numExamples / batchSize)
displayInterval = 1


```

```python
# define the functions and there derivatives

def normalizeColumnDim(weights):
    norm = np.linalg.norm(weights,axis=0)
    return(weights/(norm[None,:]+np.finfo(float).eps))
    
    

def softmax(u):
    norm = np.sum(np.exp(u),axis=0)
    return(np.exp(u)/(norm[None,:]+np.finfo(float).eps))
    
def softmaxDeriv(u):
    return(softmax(u)*(1 - softmax(u)))





def relu(u):
    return np.maximum(0,u)

def reluDeriv(u):
    uNew = np.zeros(u.shape)
    for j in range(len(u.T)):
        for i in range(len(u)):
            if relu(u[i,j]) == 0:
                uNew[i,j] = 0
            else:
                uNew[i,j] = 1
    return(uNew)
    
```

```python
def optimizeMLN(numHiddenUnits):
    # initialize weights & bias arrays
    weightsOne = np.random.randn(numInputUnits, numHiddenUnits) # first layer weights
    weightsTwo = np.random.randn(numHiddenUnits, numOutputUnits) # second layer weights
    biasOne = np.random.randn(numHiddenUnits, batchSize) # first layer bias
    biasTwo = np.random.randn(numOutputUnits, batchSize) # second layer bias
    
    # intitialize figure
    prevFig = pf.plotDataTiled(weightsOne, "First Layer Weights at Time 0", None);
    
    #fig1 = plt.figure()
    #ax1 = fig1.add_subplot(111)
    #ax1.set_xlabel('epoch')
    #ax1.set_ylabel('Training Accuracy')
    
    #fig2 = plt.figure()
    #ax2 = fig2.add_subplot(111)
    #ax2.set_xlabel('epoch')
    #ax2.set_ylabel('Error')
    
    
    etaBatch = eta/batchSize
    for t in range(2):
        # intitialize error, weights, and bias derivs
        errorT = 0
        trainAccT = 0
        deltaWeightsOne = np.zeros((numHiddenUnits, numInputUnits)) 
        deltaWeightsTwo = np.zeros((numOutputUnits, numHiddenUnits))
        deltaBiasOne = np.zeros((numHiddenUnits, batchSize))
        deltaBiasTwo = np.zeros((numOutputUnits, batchSize))
        
        for i in range(numSamples):
            # initialize data and teachers from batch
            batch = data["train"].next_batch(batchSize) 
            teacher = batch[1].T # (numOutputUnits x batchSize) dimensional array
            x = batch[0].T # (numInputUnits x batchSize) dimensional array
            
            # normalize first layer weights and bias
            weightsOne = normalizeColumnDim(weightsOne)
            biasOne = normalizeColumnDim(biasOne)
            weightsTwo = normalizeColumnDim(weightsTwo)
            biasTwo = normalizeColumnDim(biasTwo)
            
            # forward pass
            #x = dataTest[:,i,None]
            uy = (weightsOne.T @ x) + biasOne # (numHiddenUnits x batchSize) dimensional array
            y = relu(uy) # (numHiddenUnits x batchSize) dimensional array
            uz = (weightsTwo.T @ y) + biasTwo # (numOutputUnits x batchSize) dimensional array
            z = softmax(uz) # (numOutputUnits x batchSize) dimensional array
            
            
            # compute error
            #error = -np.sum(teacher * np.log(z+np.finfo(float).eps)
            #error = -np.sum((teacher * np.log(z+np.finfo(float).eps)) \
                    #+ ((1 - teacher) * np.log(1 - (z+np.finfo(float).eps))))
            
            # second layer weight & bias deriv
            #modErrorZ = (teacher / (z+np.finfo(float).eps) * softmaxDeriv(uz)
            modErrorZ = ((teacher / (z+np.finfo(float).eps)) \
                        -((1-teacher) / (1-(z+np.finfo(float).eps)))) * softmaxDeriv(uz) # (numOutputUnits x batchSize) dimensional array
        
            deltaWeightsTwo += etaBatch * (modErrorZ @ y.T) # (numOutputUnits x numHiddenUnits) dimensional array
            deltaBiasTwo += etaBatch * modErrorZ # (numOutputUnits x batchSize) dimensional array
            
            # first layer weight & bias deriv                                           
            modErrorY = (weightsTwo @ modErrorZ) * reluDeriv(uy) # (numHiddenUnits x batchSize) dimensional array
            
            deltaWeightsOne += etaBatch * (modErrorY @ x.T) # (numHiddenUnits x numInputUnits) dimensional array
            deltaBiasOne += etaBatch * modErrorY # (numHiddenUnits x batchSize) dimensional array
            
            # accumulate error
            #errorT += error
            
            # accumulate training accuracy array
            #trainAccArray = np.array(np.argmax(z,axis=0) == np.argmax(teacher,axis=0))
            #trainAccArray.astype(int)
            #trainAccT += np.sum(trainAccArray) 
               
        # update weights & bias
        weightsOne += deltaWeightsOne.T
        weightsTwo += deltaWeightsTwo.T
        biasOne += deltaBiasOne
        biasTwo += deltaBiasTwo
        
        
        
        # update figures 
        if t % 1 == 0:
            prevFig = pf.plotDataTiled(weightsOne, "First layer weights at time step "+str(t+1), prevFig)
            #ax1.scatter(t, trainAccT, s=1)
            #ax2.scatter(t,errorT)
            
    
    #prevFig = pf.plotDataTiled(weightsOne, "First layer weights at time step "+str(t), prevFig)
    
```

```python
def sigmoid(u):
    return(1/(1 + np.exp(-u))) 

def sigmoidDeriv(u):
    return(np.exp(-u)/((1 + np.exp(-u)) ** 2))
```

```python
def optimizeMLNTest(numHiddenUnits):
    # initialize weights & bias arrays
    weightsOne = np.random.randn(numInputUnits, numHiddenUnits) # first layer weights
    weightsTwo = np.random.randn(numHiddenUnits, numOutputUnits) # second layer weights
    biasOne = np.random.randn(numHiddenUnits, batchSize) # first layer bias
    biasTwo = np.random.randn(numOutputUnits, batchSize) # second layer bias
    
    # intitialize figure
    #prevFig = pf.plotDataTiled(weightsOne, "First Layer Weights at Time 0", None)
    plt.figure()
   
    etaBatch = eta/batchSize
    for t in range(100):
        # intitialize error, weights, and bias derivs
        errorT = 0
        trainAccT = 0
        deltaWeightsOne = np.zeros((numHiddenUnits, numInputUnits)) 
        deltaWeightsTwo = np.zeros((numOutputUnits, numHiddenUnits))
        deltaBiasOne = np.zeros((numHiddenUnits, batchSize))
        deltaBiasTwo = np.zeros((numOutputUnits, batchSize))
        
        for i in range(numSamples):
            # initialize data and teachers from batch
            #batch = data["train"].next_batch(batchSize) 
            #teacher = batch[1].T # (numOutputUnits x batchSize) dimensional array
            #x = batch[0].T # (numInputUnits x batchSize) dimensional array
            
            # normalize first layer weights and bias
            #weightsOne = normalizeColumnDim(weightsOne)
            #biasOne = normalizeColumnDim(biasOne)
            #weightsTwo = normalizeColumnDim(weightsTwo)
            #biasTwo = normalizeColumnDim(biasTwo)
            
            # forward pass
            x = dataTest[:,i,None]
            uy = (weightsOne.T @ x) + biasOne # (numHiddenUnits x batchSize) dimensional array
            y = sigmoid(uy) # (numHiddenUnits x batchSize) dimensional array
            uz = (weightsTwo.T @ y) + biasTwo # (numOutputUnits x batchSize) dimensional array
            z = sigmoid(uz) # (numOutputUnits x batchSize) dimensional array
            
            
            # compute error
            error = np.sum((1/2)*((teacher[:,i,None] - z)**2))
            
            # second layer weight & bias deriv
            modErrorZ = (teacher[:,i,None] - z)  * sigmoidDeriv(uz) # (numOutputUnits x batchSize) dimensional array
            deltaWeightsTwo += etaBatch * (modErrorZ @ y.T) # (numOutputUnits x numHiddenUnits) dimensional array
            deltaBiasTwo += etaBatch * modErrorZ # (numOutputUnits x batchSize) dimensional array
            
            # first layer weight & bias deriv                                           
            modErrorY = (weightsTwo @ modErrorZ) * sigmoidDeriv(uy) # (numHiddenUnits x batchSize) dimensional array
            deltaWeightsOne += etaBatch * (modErrorY @ x.T) # (numHiddenUnits x numInputUnits) dimensional array
            deltaBiasOne += etaBatch * modErrorY # (numHiddenUnits x batchSize) dimensional array
            
            # accumulate error
            errorT += error
            
            # accumulate training accuracy array
            trainAccArray = np.array(np.argmax(z) == np.argmax(teacher[:,i,None]))
            trainAccArray.astype(int)
            trainAccT += np.sum(trainAccArray) 
               
        # update weights & bias
        weightsOne += deltaWeightsOne.T
        weightsTwo += deltaWeightsTwo.T
        biasOne += deltaBiasOne
        biasTwo += deltaBiasTwo
        
        #print(t, trainAccT)
        
        
        # update figures 
        if t % 1 == 0:
            #prevFig = pf.plotDataTiled(weightsOne, "First layer weights at time step "+str(t), prevFig)
            plt.scatter(t, trainAccT,s=2)
            #plt.scatter(t,errorT,s=1)
            
            
            
    
    #prevFig = pf.plotDataTiled(weightsOne, "First layer weights at time step "+str(t), prevFig)
    
```
